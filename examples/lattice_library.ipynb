```python
import qse
import matplotlib.pyplot as plt
```

```python
def plot_lattice_2d(qbits, show_labels=False):
    _, ax = plt.subplots()

    ax.scatter(qbits.positions[:, 0], qbits.positions[:, 1])

    if show_labels:
        for i in range(qbits.nqbits):
            ax.text(qbits.positions[i, 0], qbits.positions[i, 1], i)

    ax.set_xbound(-2, 14)
    ax.set_ybound(-5, 5)
    ax.set_box_aspect(1)
```

```python
qbits = qse.lattice_library.hexagonal_lattice(1., 5, 5)
qbits.nqbits
```

```python
plot_lattice_2d(qbits, True)
plt.show()
```

```python
to_remove = [0, 49]
del qbits[to_remove]
```

```python
plot_lattice_2d(qbits, True)
plt.show()
```
